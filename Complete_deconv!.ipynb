```python
saved_peaks = d.pks
```

```python
axeZ1 = [1,2]
axeZ2 = [1,2]

full_pks_dic = {}
for z1 in axeZ1:
    for z2 in axeZ2:
        full_pks_dic["pks_{0}{1}_f1".format(z1,z2)] = []
        full_pks_dic["pks_{0}{1}_f2".format(z1,z2)] = []
        full_pks_dic["pks_{0}{1}_intensaft".format(z1,z2)] = []
        full_pks_dic["pks_{0}{1}_intensbef".format(z1,z2)] = []
        full_pks_dic["pks_{0}{1}_widthF2".format(z1,z2)] = []
        full_pks_dic["pks_{0}{1}_widthF1".format(z1,z2)] = []
for key in saved_peaks:
    for z1 in axeZ1:
        for z2 in axeZ2:
            if "Z1{0}_Z2{1}".format(z1,z2) in key:
                for e in saved_peaks[key][3]: 
                    full_pks_dic["pks_{0}{1}_f1".format(z1,z2)].append(e)
                for e in saved_peaks[key][4]:
                    full_pks_dic["pks_{0}{1}_f2".format(z1,z2)].append(e)
                for e in saved_peaks[key][2]:
                    full_pks_dic["pks_{0}{1}_intensbef".format(z1,z2)].append(e)
                for e in saved_peaks[key][5]:
                    full_pks_dic["pks_{0}{1}_intensaft".format(z1,z2)].append(e)
                for e in saved_peaks[key][7]:
                    full_pks_dic["pks_{0}{1}_widthF1".format(z1,z2)].append(e)
                for e in saved_peaks[key][8]:
                    full_pks_dic["pks_{0}{1}_widthF2".format(z1,z2)].append(e)
```

```python
import json

with open('full_deconv_pks_list.txt', 'w') as outfile:  
    json.dump(full_pks_dic, outfile)
```
